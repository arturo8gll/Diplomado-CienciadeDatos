```python
df = pd.read_csv("csv/titanic.csv")
```

```python
df["Age_cut"] = pd.cut(df["Age"], bins = 7).astype(str)
```

```python
df[["Age", "Age_cut"]]
```

```python
df["Age_cut"].value_counts(True)
```

```python
df["Embarked"] = df["Embarked"].fillna("Missing")
```

```python
df.head()
```

```python
aux = df[["Sex", "Survived"]].groupby("Sex").agg(["count", "sum"])
```

```python
aux
```

```python
aux["evento"] = aux["Survived", "sum"]
```

```python
aux["evento"]
```

```python
aux["no_evento"] = aux["Survived", "count"] - aux["Survived", "sum"]
```

```python
aux
```

```python
aux["%evento"] = aux["evento"] / aux["evento"].sum()
aux["%no_evento"] = aux["no_evento"] / aux["no_evento"].sum()
```

```python
aux
```

```python
import numpy as np
```

```python
aux["WOE"] = np.log(aux["%no_evento"] / aux["%evento"])
```

```python
aux
```

```python
aux["IV"] = (aux["%no_evento"] - aux["%evento"])*aux["WOE"]
```

```python
aux
```
